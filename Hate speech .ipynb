```python


from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Sample hateful comments for training (replace with your actual data)
hateful_comments = [
    "just by being able to tweet this insufferable bullshit proves trump a nazi you vagina",
    "denial of normal the con be asked to comment on tragedies an emotional retard",
    'me is a faggot else get out of my face',
    'i feel so much secondhand embarrassment when a white person calls me ppi or milo just say spic like you want to and move on',
]

# Sample non-hateful comments for training (replace with your actual data)
non_hateful_comments = [
    "I love the positive vibes in this community!",
    "The weather is so nice today.",
    "Great job on your project!",
    "Let's spread kindness and love.",
]

# Combine hateful and non-hateful comments
combined_comments = hateful_comments + non_hateful_comments
labels = [1] * len(hateful_comments) + [0] * len(non_hateful_comments)

# Create a DataFrame
import pandas as pd
df = pd.DataFrame({"Comments": combined_comments, "Label": labels})

# Function to preprocess text
def textProcess(text):
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    text = re.sub(r'\W', ' ', text)
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(word.lower()) for word in tokens if word.isalpha() and word.lower() not in stop_words]
    return ' '.join(tokens)

# Preprocess the comments in the DataFrame
df['processed_text'] = df['Comments'].apply(textProcess)

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(df['processed_text'])

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, df['Label'], test_size=0.2, random_state=42)

# Train Naive Bayes classifiers
mNB = MultinomialNB()
mNB.fit(X_train, y_train)

gNB = GaussianNB()
gNB.fit(X_train.toarray(), y_train)

bNB = BernoulliNB()
bNB.fit(X_train, y_train)

# Assuming you have a new text stored in a variable 'new_text'
new_text = "This is a new text that you want to classify as hateful or non-hateful."

# Preprocess the new text
processed_text = textProcess(new_text)

# Vectorize the new text using the same vectorizer
new_text_tfidf = tfidf_vectorizer.transform([processed_text])

# Make predictions using the trained classifiers
prediction_mNB = mNB.predict(new_text_tfidf)
prediction_gNB = gNB.predict(new_text_tfidf.toarray())
prediction_bNB = bNB.predict(new_text_tfidf)

# Print predictions
print("Prediction - Multinomial Naive Bayes:", prediction_mNB)
print("Prediction - Gaussian Naive Bayes:", prediction_gNB)
print("Prediction - Bernoulli Naive Bayes:", prediction_bNB)

# Check if any classifier predicts the text as hateful
is_hateful = any(prediction == 1 for prediction in [prediction_mNB, prediction_gNB, prediction_bNB])

if is_hateful:
    print(" hateful comment.")
else:
    print("non hateful comment.")
```
